```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs

x,y = make_blobs(n_samples = 40 , centers =2 , random_state=3)
# print(y)
clf = svm.SVC(kernel='linear' , C = 10)
#it must learn from the model. This is done by passing our training set to the fit method.
clf.fit(x,y)
# scatter means put dot in part that is defined
#c make the color difference
# 2 parts in first shows the dimansion
plt.scatter(x[: , 0], x [: , 1] , c =y , s = 40 )
plt.show()


# print(clf.predict([10.58,10.76]))
```

```python
w = clf.coef_[0]
print(w)

a = -w[0] / w[1]

xx = np.linspace(-6,1)
yy = a * xx - clf.intercept_[0] / w[1]

h0 = plt.plot(xx, yy, 'k-', label="linear div")

plt.scatter(x[:, 0], x[:, 1], c = y)
plt.legend()
plt.show()
```

```python
x,y = make_blobs(n_samples = 150 , centers =2 , random_state=70)
# linear ,rbf , poly 
clf = svm.SVC(kernel='rbf', C=10)
# learn from the data
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
#clf.fit(X_train, y_train)
clf.fit(x, y)

plt.scatter(x[:, 0], x[:, 1], c=y, s=30)

#find region of spot
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
# print(ax)
# print(ylim)

# # find linear diff ex : 2 3 5 [2 ,2.2,2.4,2.6,2.8,3]
# nx, ny = (3, 2)
# x = np.linspace(0, 1, nx)
# y = np.linspace(0, 1, ny)
# xv, yv = np.meshgrid(x, y)
# xv
# array([[0. , 0.5, 1. ],
#        [0. , 0.5, 1. ]])
# yv
# array([[0.,  0.,  0.],
#        [1.,  1.,  1.]])
xx = np.linspace(xlim[0], xlim[1], num =50)
yy = np.linspace(ylim[0], ylim[1], num =50)
YY, XX = np.meshgrid(yy, xx)
#ravel stick all member in multi dimansion arr to each other
# a = np.array([1, 2, 3])
# b = np.array([4, 5, 6])
# np.vstack((a,b))
# array([[1, 2, 3],
#        [4, 5, 6]])
#T , tranpose an aray 
xy = np.vstack([XX.ravel(), YY.ravel()]).T
#make shape like XX
Z = clf.decision_function(xy).reshape(XX.shape)

# level means which should u draw lines
ax.contour(XX, YY, Z, colors='k', levels = [-1,0,1]
           , linestyles=['--', '-', '--'])

ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s = 100, linewidth=1, facecolors='none')
plt.show()
```

```python
from matplotlib.image import imread
from PIL import Image as im
import os
  
# Get the path of current working directory
path = "/content/img1"  
# Get the list of all files and directories
# in current working directory
dir_list = os.listdir(path) 
# print the list
print(dir_list)
lstx = []
ylst = []
for i in range(len(dir_list)):
  try:
    mem = imread(os.path.join('img1',f"{dir_list[i]}"))
    flat = mem.flatten()
    lstx.append(flat)
    if '0_' in dir_list[i] :
      ylst.append(0)
    elif '1_' in dir_list[i] :
      ylst.append(1)
  except :
    continue


# one = imread("img/1_361.jpg")
# b  = one[: , 7:9]
# arr1 = b.reshape(2,16)
# print(arr.shape)
# print(flat)
# data = im.fromarray(b)
# saving the final output 
# # as a PNG file
# data.save('gfg_dummy_pic.png')

arrx = np.array(lstx)
print(arrx.shape)
arry = np.array(ylst)
print(arry.shape)
```

```python
# x,y = make_blobs(n_samples = 150 , centers =2 , random_state=70)
# linear ,rbf , poly 
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(arrx , arry , test_size = 0.2 , random_state = 4)
clf = svm.SVC(kernel='rbf',gamma = 'auto' ,C=10)
# learn from the data
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
# clf.fit(X_train, y_train)
clf.fit(x_train, y_train)
print(clf)
y_predict = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_predict))
```
